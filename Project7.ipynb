$$ \lnot (r \oplus p) \Rightarrow (p \lor q) $$

```python
import ttg
table = ttg.Truths(['p', 'q', 'r'], ['(p xor q)', '~(p xor q)', '(q or r)', '(~(p xor q)) => (q or r)'], ints = False)
table.as_pandas().style.set_properties(**{'text-align': 'right'}).hide_index()
```
