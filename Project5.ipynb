```python
import graphviz
gautam = graphviz.Digraph('GautamsFamilyTree')


gautam.node('Grandfather')
gautam.node('Grandmother')
gautam.edge('Grandfather', 'Deepak', label = 'son')
gautam.edge('Grandmother','Deepak',label = 'son')
gautam.edge('Deepak','Pradeep','son')
gautam.edge('Deepak','Sudeep','son')
gautam.edge('Deepak','Kuldeep','son')
gautam.edge('Pradeep','Jyoti','wife')
gautam.edge('Jyoti','Glory','daughter')
gautam.edge('Pradeep','Glory','daughter')
gautam.edge('Sudeep','Glory','niece')
gautam.edge('Kuldeep','Glory','niece')
gautam.edge('Glory','= Princess of the Family')


gautam
```
